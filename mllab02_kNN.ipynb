# 2. Normalization

> 2.1 Normalization function

2.1.1 ตัวอย่างการเขียนฟังก์ชัน normalization

````python
#nomalization method
def nomalize(X):
  new_X = []
  min_x = min(X)
  max_x = max(X)
  for x in X:
    new_X.append((x - min_x)/(max_x - min_x))

  return new_X
````

2.1.2 ตัวอย่างการเรียกใช้งาน function

````python
#example of using the normalization method
X = [100, 200, 400, 800, 1100]
new_X = nomalize(X)
print(f"input X: {X}")
print(f"Nomalized input (new_X): {new_X}")
````

2.1.3 ให้คำนวณค่า nomalize ของ input X = [200 400 600 800 1600 200 และ 2200]

````python
import matplotlib.pyplot as plt

def normalize(X):
    new_X = []
    min_x = min(X)
    max_x = max(X)
    for x in X:
        normalized_x = (x - min_x) / (max_x - min_x)
        new_X.append(normalized_x)

    return new_X

X = [200, 400, 600, 800, 1600, 200, 2200]

normalized_X = normalize(X)

# Plot the scatter plot
plt.scatter(X, normalized_X)
plt.xlabel("Original Data (X)")
plt.ylabel("Normalized Data")
plt.title("Scatter Plot of Normalized Data")
plt.show()
````

````python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

#input X of training data
x1 = [0, 1, 2, 5, 6, 7, 8]
X = list(zip(x1))

#output y of traning data
y = [0, 1, 2, 5, 6, 7, 8]

# model
neigh = KNeighborsRegressor(n_neighbors=1)
neigh.fit(X, y)

#predict
x_test = 4.5
y_hat = neigh.predict([[x_test]])

#### Plot data using matplolib ####
#plot x1, x2
plt.scatter(x1, y)
plt.xlabel("input X")
plt.ylabel("output y")

#plot prediction value
plt.scatter([x_test], y_hat, marker='*')
print(f"Prediction output for input x={x_test} is {y_hat}")

plt.show()
````

3.2.5 กำหนดให้ข้อมูลฝึกสอน

𝓣= {[x1 x2], y} = {([0 100],Red), ([0.4 300],0), ([0.6 400],1),([1.4 800],1), ([2.0 1100], 1) }

จงหาค่า nomalization ของข้อมูลอินพุต x2





```
*หมายเหตุ example 1 -> ([0 100],Red) ให้เปลี่ยน output y เป็นตัวเลข ทั้งนี้ให้ Red = 0, Blue = 1*

```


จะได้ว่า

𝓣= {[x1 x2], y} = {([0 100],0), ([0.4 300],0), ([0.6 400],1),([1.4 800],1), ([2.0 1100], 1) }



````python
import matplotlib.pyplot as plt

T = [([0, 100], 0), ([0.4, 300], 0), ([0.6, 400], 1), ([1.4, 800], 1), ([2.0, 1100], 1)]

x2_min = min(data[0][1] for data in T)
x2_max = max(data[0][1] for data in T)

normalized_x2_list = [normalization(data[0][1], x2_min, x2_max) for data in T]

for i, normalized_x2 in enumerate(normalized_x2_list):
    print(f"Normalized x2[{i+1}] = {normalized_x2:.4f}")

plt.scatter(normalized_x2_list, [data[1] for data in T], color='b', label='Data points')
plt.xlabel('Normalized x2')
plt.ylabel('y')
plt.title('Scatter Plot of T')
plt.legend()
plt.grid(True)
plt.show()

````

3.2.6 จากข้อ 3.2.5 ให้คำนายค่า output y (y_hat) เมื่อ  new input x = [0.35 350] โดยใข้ 1−NN Classifier และ 3-NN Classifier


````python
# ให้เขียน code สำหรับ k = 1
import numpy as np
import matplotlib.pyplot as plt


T = [([0, 100], 0), ([0.4, 300], 0), ([0.6, 400], 1), ([1.4, 800], 1), ([2.0, 1100], 1)]

x2_min = min(data[0][1] for data in T)
x2_max = max(data[0][1] for data in T)

def normalization(data, min_val, max_val):
    return (data - min_val) / (max_val - min_val)

# ฟังก์ชันคำนายค่า y_hat ด้วย 1-NN Classifier
def one_nn_classifier(new_input, dataset):
    normalized_new_input = normalization(new_input[1], x2_min, x2_max)
    nearest_neighbor = min(dataset, key=lambda data: abs(normalization(data[0][1], x2_min, x2_max) - normalized_new_input))
    return nearest_neighbor[1]

# new input x
x_new = np.array([0.35, 350])

# ค่า output y_hat
y_hat = one_nn_classifier(x_new, T)

print(f"Predicted output y (y_hat) for new input x: {y_hat}")

plt.scatter([normalization(data[0][1], x2_min, x2_max) for data in T], [data[1] for data in T], color='b', label='Data points')
plt.scatter(normalization(x_new[1], x2_min, x2_max), y_hat, color='r', label='New input x')
plt.xlabel('Normalized x2')
plt.ylabel('y')
plt.title('Scatter Plot with Prediction')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# ให้เขียน code สำหรับ k = 3
import numpy as np
import matplotlib.pyplot as plt

# กำหนดชุดข้อมูล T
T = [([0, 100], 0), ([0.4, 300], 0), ([0.6, 400], 1), ([1.4, 800], 1), ([2.0, 1100], 1)]

x2_min = min(data[0][1] for data in T)
x2_max = max(data[0][1] for data in T)

def normalization(data, min_val, max_val):
    return (data - min_val) / (max_val - min_val)

# ฟังก์ชันคำนายค่า y_hat ด้วย 3-NN Classifier
def three_nn_classifier(new_input, dataset):
    normalized_new_input = normalization(new_input[1], x2_min, x2_max)
    neighbors = sorted(dataset, key=lambda data: abs(normalization(data[0][1], x2_min, x2_max) - normalized_new_input))
    k_nearest_neighbors = neighbors[:3]
    y_values = [neighbor[1] for neighbor in k_nearest_neighbors]
    y_hat = max(set(y_values), key=y_values.count)  # เลือกค่า y ที่มีจำนวนมากที่สุดใน 3-nearest neighbors
    return y_hat

# new input x
x_new = np.array([0.35, 350])

# ค่า output y_hat
y_hat = three_nn_classifier(x_new, T)

print(f"Predicted output y (y_hat) for new input x: {y_hat}")

plt.scatter([normalization(data[0][1], x2_min, x2_max) for data in T], [data[1] for data in T], color='b', label='Data points')
plt.scatter(normalization(x_new[1], x2_min, x2_max), y_hat, color='r', label='New input x')
plt.xlabel('Normalized x2')
plt.ylabel('y')
plt.title('Scatter Plot with Prediction (3-NN Classifier)')
plt.legend()
plt.grid(True)
plt.show()
````

> 3.3 การใช้งาน dataset จาก sklearn library

3.3.1 ตัวอย่างการใช้ ฟังก์ขัน make_blobs ในการสร้างข้อมูลสังเคราะห์ [อ้างอิง](https://www.digitalocean.com/community/tutorials/k-nearest-neighbors-knn-in-python)

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 4)

plt.style.use('seaborn')
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], c=y, marker= '*',s=100,edgecolors='black')
plt.show()
````

3.3.2 การเตรียมข้อมูล training data และ testing data จากข้อมูลสังเคราะห์

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
````

3.3.3 การสร้าง classifier (k = 1 และ k = 5) และ ทำนายผลจากข้อมูลทดสอบ


````python
knn5 = KNeighborsClassifier(n_neighbors = 5)
knn1 = KNeighborsClassifier(n_neighbors=1)

knn5.fit(X_train, y_train)
knn1.fit(X_train, y_train)

y_pred_5 = knn5.predict(X_test)
y_pred_1 = knn1.predict(X_test)
````

3.3.4 การประเมินความแม่นยำของโมเดล

````python
from sklearn.metrics import accuracy_score

print("Accuracy with k=5", accuracy_score(y_test, y_pred_5)*100)
print("Accuracy with k=1", accuracy_score(y_test, y_pred_1)*100)
````

3.3.5 การแสดงผลข้อมูล

````python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_5, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=5", fontsize=20)

plt.subplot(1,2,2)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_1, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=1", fontsize=20)
plt.show()
````
